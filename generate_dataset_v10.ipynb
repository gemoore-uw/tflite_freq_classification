# *.mat to *.p

```python
# Note: This is compatible with data produced after 2019-10-25
from os import listdir
from os.path import isfile, join
from fnmatch import fnmatch
import scipy.io as sio
import heapq

n_data_packets_per_type = 2000
n_data_packet_types = 3
n_ctrl_packet_types = 6

n_center_freqs = 1
n_snrs = 1
n_downsamples = 3
n_mcss = 8
n_hts = 1

# Determining Maximum n_samples_per_packet so that the data tensor is large
#   enough to hold even the longest signal
n_samples_per_packet_max = 0
for filename in tqdm_notebook(sorted(listdir(data_path)),desc='Find Max # Samples Loop', leave=False):
    filepath = join(data_path, filename)
    if fnmatch(filepath, '*.mat'):
        dataDict = sio.loadmat(filepath, squeeze_me = True, struct_as_record=False)
        dataStruct = dataDict['Dataset']
        sig_list_ = np.transpose(dataStruct.signal)
        n_samples_per_packet_ = np.size(sig_list_,1)
        if n_samples_per_packet_ > n_samples_per_packet_max:
            n_samples_per_packet_max = n_samples_per_packet_
            
# n_samples_per_packet_max = 8032 # !!! Hard-coded for speed
print("Maximum # of Samples in Dataset = {}".format(n_samples_per_packet_max))

for filename in tqdm_notebook(sorted(listdir(data_path)),desc='File Open Loop', leave=False):
    filepath = join(data_path, filename)
    if fnmatch(filepath, '*.mat'):        

        # Opening the *.mat
        dataDict = sio.loadmat(filepath, squeeze_me = True, struct_as_record=False)
        dataStruct = dataDict['Dataset']        

        packet_type_list = dataStruct.packetType
        snr_list = dataStruct.snr
        n_samples_list = dataStruct.numSamples
        downsample_ratio_list = dataStruct.downsampleRate
        packet_num_list = dataStruct.packetNo
        standard = dataStruct.standard
        center_freq = dataStruct.center_freq
        
        print(filename)        
        mcs = filename.split('_')[9][-1]
        mcs2 = dataStruct.mcs
        #             print('Expected:{} - Received: {}'.format(mcs, mcs2))

        try:
            padding = dataStruct.padding
            print('Expected:{}'.format(padding))
        except AttributeError as error:
            print('AttributeError: {}'.format(error))
        except Exception as exception:
            print('Exception: {}'.format(exception))
        print('------------------------------------')
```
